```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def tmp(tmax, tratio, k, nb_iter):
    return tmax * np.exp(tratio * k / nb_iter)
```

```python
def acc(t, deltae):
    return np.exp(- deltae / t)
```

```python
tmax = 1
tmin = 0.5
nb_iterr = 10000000
tratio = -np.log(tmax / tmin)

k = np.linspace(0, nb_iterr, 50)

i = 0
plt.figure(figsize = (14, 9))
for delta in [2, 4, 6, 8]:
    
    t = tmp(tmax, tratio, k, nb_iterr)
    a = acc(t, delta)
    plt.subplot(2, 2, i + 1)
    plt.plot(k, tmp(tmax, tratio, k, nb_iterr), label="Temperature")
    plt.plot(k, a, label=f'Acceptance probability for $\Delta$ = {delta}')
    plt.ylim(0, 1)
    plt.xlim(0, nb_iterr)
    plt.legend()
    plt.xlabel('Iterations')
    i += 1
plt.show()
```
